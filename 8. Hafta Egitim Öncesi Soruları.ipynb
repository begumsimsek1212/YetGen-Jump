# Problem 1 
Kullanıcıdan aldığı ürün bilgisini (ad, fiyat) urunler.txt dosyasına kayıt eden fonksiyonu yazınız.

````python
def urun_kaydet(ad, fiyat):
    with open("urunler.txt", "a") as f:
        f.write(f"{ad} {fiyat}\n")
````

# Problem 2
1.problemde oluşturulan text dosyası içerisinde kullanıcıdan alacağımız eski kelime isimli değişkene atadığımız 
ve değiştirilmesini istediğimiz kelimemizi yerini alıcak olan yeni kelime isimli değişkene atayın. Belirlenen kelimeleri text dosyasında değiştiren fonksiyonu yazınız.

````python
def kelime_degistir(urunler.txt , eski_kelime, yeni_kelime):
    with open(urunler.txt, "r") as f:
        a = f.read()
    a = a.replace(eski_kelime, yeni_kelime)
    with open(urunler.txt, "w") as f:
        f.write(a)
````

# Problem 3
Vehicle sınıfının tüm değişkenlerini ve yöntemlerini devralacak bir alt sınıf olarak Bus sınıfı oluşturunuz.



```
Output
> Name: School Volvo Speed: 180 Mileage: 12
```


````python
class Vehicle:
    def __init__(self, make, model, year, weight):
        self.make = make
        self.model = model
        self.year = year
        self.weight = weight
    
    def start(self):
        print("The vehicle is starting.")
    
    def stop(self):
        print("The vehicle is stopping.")
    
    def drive(self):
        print("The vehicle is driving.")

class Bus(Vehicle):
    def __init__(self, make, model, year, weight, capacity):
        super().__init__(make, model, year, weight)
        self.capacity = capacity

def announce(self):
        print(f"This is a {self.make} {self.model} bus built in {self.year} with a weight of {self.weight} kg and a capacity of {self.capacity} passengers.")
````
